```python
# Importing the relevant libraries for the analysis

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("C:/Users/Mr Lekan/OneDrive/Desktop/CoffeeShop.xlsx")
df.head()
```

## PROCESS
This is where the dataset is scrubbed for analysis stage. I will check and correct the following:
* missing values
* duplicated values
* wrong format
* wrong values


```python
# Finding and Correcting Missing Values
missing_values = df.isnull().sum()
print(missing_values)
```

```python
df.info()
```

No missing values

```python
# Finding Duplicates and deleting them

df.duplicated().sum()
```

No duplicated values

```python
# Converting transation_time to timedelta for easy manipulation

df["transaction_time"] = df["transaction_time"].astype(str)
df["transaction_time"] = pd.to_timedelta(df["transaction_time"])
```

```python
# Creating Sales and datetime columns

df["sales"]= df["transaction_qty"] * df["unit_price"]
df["datetime"] = df["transaction_date"] + df["transaction_time"]
df.head(5)
```

## Analyze

```python
df.describe()
```

```python
# q  Grouping sales and number of transaction by store_locations 
```

```python
# hourly sales by product category

df["hour"] = df["datetime"].dt.hour
hourly_sales_by_product = df.groupby(["hour", "product_category"])["sales"].sum().unstack()
hourly_sales_by_product.plot(kind="bar", stacked=True, figsize=(14,8), colormap="viridis")
plt.xlabel("Hour of the Day")
plt.ylabel("Total Sales ($)")
plt.title("Hourly Sales by Product Category")
plt.legend(title="Product Category", loc="upper left")
plt.grid(True)
plt.tight_layout()

plt.show()
```

Largest sales was recorded by 10am, and least sales was recorded by 8pm.

```python
# Daily Sales by Store Location

daily_sales_by_locations = df.groupby(["transaction_date", "store_location"])["sales"].sum().unstack()
daily_sales_by_locations
```

```python
daily_sales_by_locations.plot(figsize= (14, 8), title = " Daily Sales by Store Locations")
plt.xlabel("Date")
plt.ylabel("Total Sales ($)")
plt.legend(title="Store Locations")
plt.grid(True)
plt.xticks(rotation= 45)
plt.tight_layout()

plt.show()
```

```python
# Looking at the same data in Stacked Area Chart

plt.figure(figsize= (14, 8))
plt.stackplot(daily_sales_by_locations.index, daily_sales_by_locations.T, labels=daily_sales_by_locations.columns)
plt.title("Daily Sales by Store Locations (Stacked Area Chart)")
plt.xlabel("Date")
plt.ylabel("Total Sales ($)")
plt.legend(title="Store Locations", loc="upper left")
plt.grid(True)
plt.xticks(rotation= 45)
plt.tight_layout()

plt.show()
```

Across the three locations, daily sales were in ascending order from January to June.

```python
# Weekly Sales by Locations

weekly_sales = df.groupby(["day_of_week","store_location"])["sales"].sum().unstack()
days_ordered= ["Monday","Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
weekly_sales = weekly_sales.reindex(days_ordered)
weekly_sales
```

```python
weekly_sales[["Astoria","Hell's Kitchen","Lower Manhattan"]].plot(kind="bar", figsize=(14,8))

plt.xlabel("Days of the Week", fontweight='bold')
plt.ylabel("Total Sales ($)", fontweight="bold")
plt.title("Weekly Sales by Locations")
plt.legend(title="Store Locations", loc="upper left")

plt.tight_layout()
plt.show()
```

```python
# Orderly Weekly Sales 

weekly_trans = df.groupby("day_of_week")["transaction_id"].count()
days_ordered= ["Monday","Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
weekly_trans = weekly_trans.reindex(days_ordered).sort_values(ascending=False)
weekly_trans
```

```python
plt.style.use("fivethirtyeight")
weekly_trans.plot(kind="barh", figsize=(12,6))
plt.xlabel("No of Transactions", fontweight="bold")
plt.ylabel("Day of Week", fontweight="bold")
plt.title("Transactions by Day of Week")


plt.tight_layout()
plt.show()
```

Monday has the largest sales, followed by Friday. Least sales was recorded on Saturday.

```python
# Monthly Sales

df["month"] = df["datetime"].dt.to_period("M")
monthly_sales = df.groupby("month")["sales"].sum().reset_index()

monthly_sales
```

```python
monthly_sales.plot(figsize=(14,8), title="Monthly Sales")
plt.xlabel("Month")
plt.ylabel("Monthly Sales ($)")
plt.xticks(rotation= 45)
plt.tight_layout()

plt.show()
```

Highest sales was recorded in June while the lowest February has the lowest sales figure.

```python
# Looking at product category that sold most and least

product_category = df.groupby("product_category")["transaction_id"].count().sort_values(ascending=False)

product_category
```

```python
product_category.plot(kind="bar", stacked=True, figsize=(12,6), colormap="viridis")
plt.xlabel("Product Category")
plt.ylabel("No of Transactions")
plt.title(" Transactions by Product Category")

plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

Coffee category has the highest numbers of transactions while the least is Packaged Chocolate.

```python
# product type that sold most and least

sales_by_product = df.groupby("product_type")["transaction_id"].sum().sort_values(ascending=False)

sales_by_product
```

```python
sales_by_product.plot(kind="bar", stacked=True, figsize=(12,6), colormap="viridis")
plt.xlabel("Product Type")
plt.ylabel("No of Transactions")
plt.title("Transactions by Product Type")

plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

By type, the product that sold most is Brewed chai tea while Green beans has the least number of sales.

```python
# Revenue contributions product category 

product_category = df.groupby("product_category")["sales"].sum().sort_values(ascending=False)

product_category
```

```python
product_category.plot(kind='pie', figsize=(14,8), autopct = '%1.1f%%')
plt.title('Percentage Sales by Product Category')
plt.axis('equal')
plt.legend(title="Product Category", loc='upper left')

plt.show()
```

Coffee category contributed most to revenue while Packaged Chocolate category made the least revenue

```python
# Revenue contribution by product type

rev_by_product_type = df.groupby("product_type")["sales"].sum().sort_values(ascending=False)

rev_by_product_type
```

```python
rev_by_product_type.plot(kind="bar", stacked=True, figsize=(12,6), colormap="viridis")
plt.xlabel("Product Type", fontweight='bold')
plt.ylabel("Tota Sales ($)", fontweight="bold")
plt.title("Revenue Conribution by Product Type")

plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

By product type, Barista Espresso turned in most revenue while the least revenue came from Green beans.

```python
# Product type sales under each category

product_sales = df.groupby(["product_category", "product_type"])["sales"].sum().reset_index()
product_sales
```

```python
plt.figure(figsize=(14,8))
sns.barplot(data=product_sales, y="sales", x="product_category", hue="product_type")

plt.xlabel("Product Category")
plt.ylabel("Total Sales ($)")
plt.title("Product Sales by Category and Type")
plt.legend(title='Product Type', loc="upper right")

plt.tight_layout()
plt.show()
```

```python
# Making it clearer below
```

```python
categories = product_sales["product_category"].unique()

for category in categories:
    plt.figure(figsize=(5,5))
    category_data = product_sales[product_sales["product_category"]==category]
    sns.barplot(data=category_data, x="product_type", y="sales")
    
    # Add labels and title
    plt.title(f'Sales by Product Type in {category}')
    plt.xlabel("Product Type")
    plt.ylabel("Total Sales ($)")
    plt.xticks(rotation=45)
    
    plt.tight_layout()
    plt.show()
```

```python
# Finding product category by location

product_category_by_locations = df.groupby(["store_location","product_category"])["transaction_id"].count().unstack()
stores_ordered = ["Astoria", "Hell's Kitchen", "Lower Manhattan"]
product_category_by_locations=product_category_by_locations.reindex(stores_ordered)
product_category_by_locations

```

```python
plt.style.use("ggplot")
product_category_by_locations.plot(kind="bar", figsize=(14,8))

plt.title("Product Category Transactions by Locations")
plt.ylabel("No of Transaction", fontweight="bold")
plt.xlabel("Store Locations", fontweight="bold")
plt.legend(title="Product Category", loc="upper left")
plt.grid(True)
plt.tight_layout()
plt.xticks(rotation=45)

plt.show()
```

Coffee catgory was mostly sold across the locations. Packaged Chocolate had the least number of transactions.

```python
# Over Contribution of Store Locations to Sales

percent_sales_by_locations= df.groupby("store_location")["sales"].sum()
percent_sales_by_locations.plot(kind='pie', figsize=(14,4), autopct = '%1.1f%%')
plt.title('Percentage Sales by Locations')
plt.axis('equal')
plt.legend(title="Store Locations", loc='upper left')

plt.show()
```

although very close to one another, Hell's Kitchen made the largest revenue followed by Astoria.
